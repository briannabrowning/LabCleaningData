```python
# Question 1: Print out the Outliers in each column in the pima_fixed_columns_df dataframe
# use the IQR method of outlier detection.

def is_outlier(pima_fixed_columns_df):
    Q1 = pima_fixed_columns_df.quantile(.25)
    Q3 = pima_fixed_columns_df.quantile(.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    return pima_fixed_columns_df[~pima_fixed_columns_df.between(lower_bound, upper_bound)]

```
